```python
import pandas as pd

df1 = pd.read_csv("BAI Continuous Tracker_Respondent_Level_Data_01.csv", encoding="cp1252")
df1
```

```python
filter_col = [col for col in df1 if col.startswith("Q50_") or col.startswith("Q60_")]
filter_col.append("DUMEndDate")
filter_col
df1_mod = df1[filter_col]
df1_mod.dropna(axis=0, how="any", inplace=True)
df1_mod
```

```python
import numpy as np

mapping = {"Q50_01": ["Q60_01", "Q60_02", "Q60_03", "Q60_04"], "Q50_02": ["Q60_05", "Q60_06"], "Q50_03": ["Q60_07"], 
           "Q50_04": ["Q60_08", "Q60_09", "Q60_10"], "Q50_05": ["Q60_11", "Q60_12", "Q60_13"], "Q50_06": ["Q60_14"], 
           "Q50_07": ["Q60_15"], "Q50_08": ["Q60_16", "Q60_17"], "Q50_09": ["Q60_18"], "Q50_10": ["Q60_19"], 
           "Q50_11": ["Q60_20", "Q60_21"], "Q50_12": ["Q60_22"], "Q50_13": ["Q60_23", "Q60_24"], "Q50_14": ["Q60_25"],
           "Q50_15": ["Q60_26", "Q60_27"], "Q50_16": ["Q60_28"], "Q50_17": ["Q60_29", "Q60_30"], "Q50_18": ["Q60_31", "Q60_32"],
           "Q50_19": ["Q60_33"], "Q50_20": ["Q60_34"], "Q50_21": ["Q60_35"], "Q50_22": ["Q60_36", "Q60_37"],
           "Q50_23": ["Q60_38"], "Q50_24": ["Q60_39"], "Q50_25": ["Q60_40", "Q60_41"], "Q50_26": ["Q60_42", "Q60_43"],
           "Q50_27": ["Q60_44", "Q60_45"], "Q50_28": ["Q60_46", "Q60_47"], "Q50_29": ["Q60_48", "Q60_49"], "Q50_30": ["Q60_50"],
           "Q50_31": ["Q60_51", "Q60_52"], "Q50_32": ["Q60_53"], "Q50_33": ["Q60_54", "Q60_55"], "Q50_34": ["Q60_56", "Q60_57"],
           "Q50_35": ["Q60_58"], "Q50_36": ["Q60_59", "Q60_60"], "Q50_37": ["Q60_61"], "Q50_38": ["Q60_62", "Q60_63"],
           "Q50_39": ["Q60_64", "Q60_65"]}

for k, v in mapping.items():
    try:
        col_name = "Aware_" + k.split("_")[1]
        df1_mod[col_name] = np.nan
        for index, row in df1_mod.iterrows():
            aware_sum = 0
            for item in v:
                aware_sum += row[item]
        
            if row[k] == 1 or aware_sum > 0:
                df1_mod.at[index, col_name] = 1
            else:
                df1_mod.at[index, col_name] = 0
    except:
        print ("error at ", k)
        continue

df1_mod[["Aware_01", "Aware_02"]]
```

```python
from datetime import datetime, timedelta

def get_week_start(date):
    return date - timedelta(days=date.weekday()+1)

def get_week_end(date):
    return get_week_start(date) + timedelta(days=6)

df1_mod.dropna(axis=1, how="all", inplace=True)
df1_mod["Aware_Bai"] = df1_mod["Aware_01"]
filter_col = [col for col in df1_mod if col.startswith("Aware_")]
filter_col.remove("Aware_01")
filter_col.remove("Aware_Bai")
df1_mod["Date"] = df1_mod["DUMEndDate"].apply(lambda x: x.split()[0])
df1_mod["Date"] = df1_mod["Date"].apply(lambda x: datetime.strptime(x, "%m/%d/%Y"))

df1_mod["Aware_Others"] = df1_mod[filter_col].mean(axis=1)
df1_mod["Week_Start"] = df1_mod["Date"].apply(lambda x: get_week_start(x))
df1_mod["Week_End"] = df1_mod["Date"].apply(lambda x: get_week_end(x))
df1_mod

#Formula to calculate week starting Monday and ends Sunday
#start = dt - timedelta(days=dt.weekday())
#end = start + timedelta(days=6)
df1_final = df1_mod[["Aware_Bai", "Aware_Others", "Date", "Week_Start", "Week_End"]]
df1_final = df1_final.groupby(["Week_Start", "Week_End"]).agg({"Aware_Bai": "mean", "Aware_Others": "mean"})
df1_final.reset_index(inplace=True)
df1_final.sort_values("Week_Start", inplace=True)
df1_final
```
